# Method-1

```python
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x,y)
plt.show()
```

# Method-2

```python
fig  =  plt.figure()
ax = fig.add_axes([1,1,1,1])
ax.plot(x,y)
plt.show()
```

# Method-3


```python
# 1-prepare the data
x = [1,2,3,4,5,6,7]
y = [12,34,67,12,45,12,45]
# 2- Set a plot envirnment
home, room = plt.subplots()
# 3-plot the graph
room.plot(x,y)
# 4-Customization
room.set(title='A Good Plot',
         xlabel='x-axis',
         ylabel='y-axis',
)
plt.show()
```

```python
# create figure and graph area
fig, ax = plt.subplots(figsize=(10,6))
#home, room
# plot the data 
scatter_plot = ax.scatter(x=over_40['age'],
                         y=over_40['chol'],
                         c=over_40['target'])
# customization
ax.set(title='Suspectability of Heart Diseaser',
      xlabel='Age',
      ylabel='cholesterol')

# set the legend
ax.legend(*scatter_plot.legend_elements(), title='target')
scatter_plot;
```

# Entire Life Cycle of Object Oriented Plot Advanced

```python
#1 Create Plot area

fig,(plot1,plot2) = plt.subplots(nrows=2,
                                 ncols=1,
                                 figsize=(10,10),
                                sharex=True)
#2 Add data to plot1
scatter_plot = plot1.scatter(x=over_40['age'],
                            y=over_40['chol'],
                            c=over_40['target'],
                            cmap='winter',
                            )
#3 customize
plot1.set(title='Heart Disease Problem with Cholesterol',
         #xlabel='Age',
         ylabel='Cholesterol',)
plot1.legend(*scatter_plot.legend_elements(), title='Survived or Not')
plot1.axhline(y=over_40['chol'].mean(),linestyle="--");
######Now lets Plot Second Plot
scatter_plot = plot2.scatter(x=over_40['age'],
                            y=over_40['trestbps'],
                            c=over_40['target'],
                             cmap='plasma'
                            )
#4 Customize
plot2.set(title='Heart Disease Problem with resting Blood Pressure',
         xlabel='Age',
         ylabel='Resting Blood Pressure',)
#5 Add Legend
plot2.legend(*scatter_plot.legend_elements(),title='Survived or Not');
#6 Meanline
plot2.axhline(y=over_40['trestbps'].mean(),linestyle="--");
#7 coloring data
#8 Coloring plots
plt.style.use('seaborn-darkgrid')
#9 Naming the Figure
fig.suptitle('Patients Data',fontsize=32,fontweight='bold');
#10 Saving the figrue
fig.savefig('Patients data report.png')
```
